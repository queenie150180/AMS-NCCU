# 資料分析 : 104學年度大專院校校別學生數

```python
df1 = pd.read_csv("http://stats.moe.gov.tw/files/detail/104/104_student.csv")
```

```python
df1.head()
```

```python
len(df1[df1.學校代碼 == 1])
```

```python
df1.sort_values(by="一年級男生",ascending=False).head()
```

```python
df1.sort_values(by="一年級女生",ascending=False).head()
```

```python
df1.insert(6, column = "一年級學生數", value = df1.一年級男生 + df1.一年級女生 )
```

```python
df1.head()
```

```python
df1.insert(9, column = "二年級學生數量", value = df1.二年級男生 + df1.二年級女生 )
df1.insert(12, column = "三年級學生數量", value = df1.三年級男生 + df1.三年級女生 )
df1.insert(15, column = "四年級學生數量", value = df1.四年級男生 + df1.四年級女生 )
```

```python
df1.drop("五年級男生", axis=1, inplace=True)
```

```python
df1.drop("五年級女生", axis=1, inplace=True)
df1.drop("六年級男生", axis=1, inplace=True)
df1.drop("六年級女生", axis=1, inplace=True)
df1.drop("七年級男生", axis=1, inplace=True)
df1.drop("七年級女生", axis=1, inplace=True)
df1.drop("延修生男生", axis=1, inplace=True)
df1.drop("延修生女生", axis=1, inplace=True)
```
