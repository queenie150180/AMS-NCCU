## 重量轉換

```python
a = input("請輸入重量(公斤):")
a = float(a)
b = a*1.66
print("你輸入的",a,"公斤 = ",'%.2f'% b,"台斤")
```

## 圖形

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (13 * np.random.rand(N))**2  
plt.scatter(x, y, area, colors, alpha=0.65);
```

## 看劇選擇

```python
import random
```

```python
Comedy = ["大力女子金奉順","三流之路","來自星星的你","她很漂亮","Dream High","孤單又燦爛的神－鬼怪"]
Romance = ["喜歡的話請響鈴","皮諾丘","藍色海洋的傳說","舉重妖精金福珠","焦急的羅曼史"]
Suspense = ["檢法男女2","三年A班","輪到你了","Signal","W－兩個世界"]
Other = ["天空之城","請輸入檢索詞:WWW","醫生耀漢","派遣女醫","多謝款待"]

msg = "追劇追起來"

while (msg != "想睡了"):
    msg = input("今天心情如何?")
    if "難過" in msg:
        print("那就看一點好笑的吧! 我推薦你看:",random.choice(Comedy))
    elif "戀愛" in msg:
        print("那就看一下帥氣的歐爸吧! 我推薦你看:",random.choice(Romance))
    elif "懸疑" in msg:
        print("那來看懸疑片動動大腦! 我推薦你看:",random.choice(Suspense))
    elif "普通" in msg:
        print("那我推薦你: ",random.choice(Other))
    else:
        print("那就睡覺吧!ㄅㄅ")
```

## 天氣又開始熱了要防紫外線

```python
print("貼心小精靈幫你做好防曬ξ( ✿＞◡❛)")

a = input("請輸入紫外線指數:")
a = int(a)
if 0<=a<=2:
    print("指標為:","\033[7;32m 綠色 \033[0m","對人體沒有危害，如有需要配戴太陽眼鏡即可!")
elif 3<=a<=5:
    print("指標為:","\033[7;33m 黃色 \033[0m","小叮嚀:記得要擦防曬乳了唷!")
elif 6<=a<=10:
    print("指標為:","\033[7;31m 紅色 \033[0m","這個太陽真的不行了!要出去一定要全套防護!")
elif a>=11:
    print("指標為:","\033[7;35m 紫色 \033[0m","紫爆了紫爆了!!!躲在室內不要出去(((ﾟДﾟ;)))")
```
